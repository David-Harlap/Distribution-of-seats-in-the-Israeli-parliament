### The function 

define function that recive lambda function, number that represents the number of seats to be divided and the name of the column which we will store the information. 

```python
def divisor_method(funct, seat_to_give,name, df):
    df[name]=0
    for i in range(seat_to_give):
        df['temp'] = df['people'] / funct(df[name])
        df[name][df['temp'].idxmax()]+=1
```

### test my function with class example

```python
ex = {'Names':['A', 'B', 'C'], 'people':[210,50,40]}
test_df = pd.DataFrame(ex)
```

```python
test_df
```

```python
divisor_method((lambda a : a+1),3,'jefferson_seat', test_df )
divisor_method((lambda a : a+0.5),3,'webster_seat', test_df )
divisor_method((lambda a : a),3,'adams_seat',test_df )
```

```python
test_df= test_df.drop(axis=1, columns='temp')
test_df
```
